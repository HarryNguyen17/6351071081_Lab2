```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# 1. Đọc & tiền xử lý
data = pd.read_csv("heart.csv").dropna()
data.columns = [
    "age","sex","chest_pain_type","resting_blood_pressure","cholesterol",
    "fasting_blood_sugar","rest_ecg","max_heart_rate_achieved",
    "exercise_induced_angina","st_depression","st_slope","num_major_vessels",
    "thalassemia","target"
]

# 2. Tách features / labels
features = data.drop("target", axis=1)
labels = data["target"]

# 3. One-hot các cột không phải dạng số
non_numeric_cols = features.select_dtypes(exclude=["int64", "float64"]).columns
features_onehot = pd.get_dummies(features, columns=non_numeric_cols, drop_first=True)

# 4. Tách train / test (80 – 20)
train_size = int(len(features_onehot) * 0.8)

X_train = features_onehot[:train_size]
X_test  = features_onehot[train_size:]

y_train = labels[:train_size]
y_test  = labels[train_size:]

# 5. Xây dựng cây ID3 (entropy)
model = DecisionTreeClassifier(criterion="entropy", max_depth=None)
model.fit(X_train, y_train)

# 6. Kiểm thử → Confusion Matrix
y_pred = model.predict(X_test)

cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap="Blues")
plt.title("Confusion Matrix - ID3 (Entropy)")
plt.show()

# 7. Biểu diễn cây
plt.figure(figsize=(18, 10))
plot_tree(
    model,
    feature_names=X_train.columns,
    class_names=["No disease", "Disease"],
    filled=True,
    rounded=True,
    fontsize=7
)
plt.title("Decision Tree (ID3 - Entropy)")
plt.show()

print("Nguyễn Đình Vương - 6351071081\n")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# 1. Đọc & tiền xử lý
data = pd.read_csv("heart.csv").dropna()
data.columns = [
    "age","sex","chest_pain_type","resting_blood_pressure","cholesterol",
    "fasting_blood_sugar","rest_ecg","max_heart_rate_achieved",
    "exercise_induced_angina","st_depression","st_slope","num_major_vessels",
    "thalassemia","target"
]

# 2. Tách features / labels
features = data.drop("target", axis=1)
labels = data["target"]

# 3. One-hot các cột không phải dạng số
non_numeric_cols = features.select_dtypes(exclude=["int64", "float64"]).columns
features_onehot = pd.get_dummies(features, columns=non_numeric_cols, drop_first=True)

# 4. Tách train / test (80 – 20)
train_size = int(len(features_onehot) * 0.8)

X_train = features_onehot[:train_size]
X_test  = features_onehot[train_size:]

y_train = labels[:train_size]
y_test  = labels[train_size:]

# 5. Xây dựng cây CART (Gini)
cart_model = DecisionTreeClassifier(criterion="gini", max_depth=None)
cart_model.fit(X_train, y_train)

# 6. Kiểm thử → Confusion Matrix
y_pred_cart = cart_model.predict(X_test)

cm_cart = confusion_matrix(y_test, y_pred_cart)
disp = ConfusionMatrixDisplay(confusion_matrix=cm_cart)
disp.plot()
plt.title("Confusion Matrix - CART (Gini)")
plt.show()

# 7. Biểu diễn cây CART
plt.figure(figsize=(18, 10))
plot_tree(
    cart_model,
    feature_names=X_train.columns,
    class_names=["No disease", "Disease"],
    filled=True,
    rounded=True,
    fontsize=7
)
plt.title("Decision Tree - CART (Gini)")
plt.show()

print("Nguyễn Đình Vương - 6351071081\n")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay

# 1. Đọc & tiền xử lý
data = pd.read_csv("heart.csv").dropna()
data.columns = [
    "age","sex","chest_pain_type","resting_blood_pressure","cholesterol",
    "fasting_blood_sugar","rest_ecg","max_heart_rate_achieved",
    "exercise_induced_angina","st_depression","st_slope","num_major_vessels",
    "thalassemia","target"
]

# 2. Tách features / labels
features = data.drop("target", axis=1)
labels = data["target"]

# 3. One-hot cho các cột không phải số
non_numeric_cols = features.select_dtypes(exclude=["int64", "float64"]).columns
features_onehot = pd.get_dummies(features, columns=non_numeric_cols, drop_first=True)

# 4. Tách train / test (80 – 20, giống các câu trước)
train_size = int(len(features_onehot) * 0.8)

X_train = features_onehot[:train_size]
X_test  = features_onehot[train_size:]

y_train = labels[:train_size]
y_test  = labels[train_size:]

# 5. Xây dựng mô hình Naïve Bayes (GaussianNB)
nb_model = GaussianNB()
nb_model.fit(X_train, y_train)

# 6. Dự đoán và đánh giá
y_pred = nb_model.predict(X_test)

acc = accuracy_score(y_test, y_pred)
print("Accuracy (Naïve Bayes):", acc)

cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.title("Confusion Matrix - Naïve Bayes (GaussianNB)")
plt.show()

print("Nguyễn Đình Vương - 6351071081\n")
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# 1. Đọc & xử lý như trước
data = pd.read_csv("heart.csv").dropna()
data.columns = [
    "age","sex","chest_pain_type","resting_blood_pressure","cholesterol",
    "fasting_blood_sugar","rest_ecg","max_heart_rate_achieved",
    "exercise_induced_angina","st_depression","st_slope","num_major_vessels",
    "thalassemia","target"
]

features = data.drop("target", axis=1)
labels = data["target"]

non_numeric_cols = features.select_dtypes(exclude=["int64", "float64"]).columns
features_onehot = pd.get_dummies(features, columns=non_numeric_cols, drop_first=True)

# 80% train / 20% test
train_size = int(len(features_onehot) * 0.8)
X_train = features_onehot[:train_size]
X_test  = features_onehot[train_size:]
y_train = labels[:train_size]
y_test  = labels[train_size:]

# 2. Train các mô hình
id3 = DecisionTreeClassifier(criterion="entropy")
id3.fit(X_train, y_train)

cart = DecisionTreeClassifier(criterion="gini")
cart.fit(X_train, y_train)

nb = GaussianNB()
nb.fit(X_train, y_train)

# 3. Dự đoán & tính accuracy
acc_id3  = accuracy_score(y_test, id3.predict(X_test))
acc_cart = accuracy_score(y_test, cart.predict(X_test))
acc_nb   = accuracy_score(y_test, nb.predict(X_test))

# 4. In kết quả so sánh
print("=== KẾT QUẢ SO SÁNH ===")
print(f"ID3 Accuracy : {acc_id3:.4f}")
print(f"CART Accuracy: {acc_cart:.4f}")
print(f"Naive Bayes  : {acc_nb:.4f}")

print("Nguyễn Đình Vương - 6351071081\n")
```
